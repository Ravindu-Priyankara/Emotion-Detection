```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
from keras.models import load_model
import numpy as np
```

```python
facedetect= cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cap = cv2.VideoCapture(1)
cap.set(3,640)
cap.set(4,480)
font=cv2.FONT_HERSHEY_COMPLEX
```

```python
model = load_model('keras_model.h5')
```

```python
while True:
    success,imOrignal = cap.read()
    faces = facedetect.detectMultiScale(imOrignal,1.3,5)

    for x,y,w,h in faces:
        crop_img =imOrignal[y:y+h,x:x+h]
        img= cv2.resize(crop_img, (224,224))
        img = img.reshape(1,224,224,3)
        prediction = model.predict(img)
    
        #classIndex= model.predict_classes(img)
        probabilityValue = np.amax(prediction)
        print(type(probabilityValue))
        
        
        org = (150, 50)
        
        if float(probabilityValue) < -1:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0), 2)
            cv2.rectangle(imOrignal,(x,y-40),(x+w, y),(0,255,0),-2)
            cv2.putText(imOrignal,'Angry',org,font,0.75,(255,255,255),1,cv2.LINE_AA)

            file = open("data.txt","a")
            file.write(str(probabilityValue))
            file.write("\n")
            file.close()

        elif float(probabilityValue)>1.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'Disgust',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
        
        elif float(probabilityValue)>2.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'fear',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
            
        elif float(probabilityValue)>3.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'happy',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
        
        elif float(probabilityValue)>4.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'neutral',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
        
        elif float(probabilityValue)>5.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'sad',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
        
        elif float(probabilityValue)>6.0:
            cv2.rectangle(imOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imOrignal,'suprise',org, font, 0.75, (255,255,255),1, cv2.LINE_AA)
        cv2.putText(imOrignal,str(round(probabilityValue*100, 2))+"%" ,(180, 75), font, 0.75, (255,0,0),2, cv2.LINE_AA)
        
        
    cv2.imshow("Result",imOrignal)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
